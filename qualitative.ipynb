```python
import numpy as np
import json
from evaluation import most_similar

embeddings = np.load("embeddings.npy")
with open("word_to_id.json") as f:
    word_to_id = json.load(f)
id_to_word = {i: word for word, i in word_to_id.items()}
```

```python
vec = most_similar(embeddings[word_to_id["king"]], embeddings, exclude=[word_to_id["king"]])
print(id_to_word[vec]) # anshan - some city
vec = most_similar(embeddings[word_to_id["king"]], embeddings, exclude=[word_to_id["king"], word_to_id["anshan"]])
print(id_to_word[vec]) # canute - king of England, Denmark, Norway
vec = most_similar(embeddings[word_to_id["king"]], embeddings, exclude=[word_to_id["king"], word_to_id["anshan"], word_to_id["canute"]])
print(id_to_word[vec]) # lulach - king of Scotland
vec = most_similar(embeddings[word_to_id["king"]], embeddings, exclude=[word_to_id["king"], word_to_id["anshan"], word_to_id["canute"], word_to_id["lulach"]])
print(id_to_word[vec]) # aedh - many kings with this name
```

```python
def top_n(vec, embeddings, id_to_word, n=5, exclude=None):
    norms = np.linalg.norm(embeddings, axis=1)
    scores = embeddings @ vec / (norms * np.linalg.norm(vec))
    if exclude:
      for idx in exclude:
          scores[idx] = -np.inf
    top = np.argsort(scores)[-n:][::-1]
    return [(id_to_word[i], scores[i]) for i in top]
```

```python
vec = embeddings[word_to_id["king"]] - embeddings[word_to_id["man"]] + embeddings[word_to_id["woman"]]
# Does not return "queen", but context is good
print(top_n(vec, embeddings, id_to_word, n=10, exclude=[word_to_id["king"], word_to_id["man"], word_to_id["woman"]]))
```
